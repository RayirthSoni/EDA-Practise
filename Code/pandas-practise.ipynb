```python
import pandas as pd
import numpy as np

# Seed for reproducibility
np.random.seed(42)

# Generate data
order_id = np.arange(1001, 1051)
customer_id = np.random.randint(100, 200, size=50)
order_date = pd.date_range(start="2023-01-01", periods=50, freq="D")
product_id = np.random.randint(1000, 1010, size=50)
quantity = np.random.randint(1, 10, size=50)
price_per_unit = np.random.uniform(10, 100, size=50).round(2)
discount = np.random.uniform(0, 0.3, size=50).round(2)
total_price = (quantity * price_per_unit * (1 - discount)).round(2)

# Create DataFrame
data = pd.DataFrame(
    {
        "order_id": order_id,
        "customer_id": customer_id,
        "order_date": order_date,
        "product_id": product_id,
        "quantity": quantity,
        "price_per_unit": price_per_unit,
        "discount": discount,
        "total_price": total_price,
    }
)

data
```
